## 导入绘图库

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 设置区间大小和计算函数

```python
x_range = [i for i in range(50)]  # 列表生成式
x = np.array(x_range)
y = 2 * x + 1
```

## 绘图

```python
plt.figure(1)
plt.plot(x,y,linewidth = 4.0,color = 'r')  #细节调整的两个方式
plt.title("y = 2 * x + 1", fontsize = 18)  #设置标题
plt.xlabel("x(n)",fontsize = 14)  #对横纵轴进行说明
plt.ylabel("y(n)",fontsize = 14)  
plt.show()
```
