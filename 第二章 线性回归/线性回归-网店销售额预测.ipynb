# 数据可视化
首先需要进行数据可视化工作，看看数据大概长什么样子

```python
# 导入必要的库
import numpy as np
import pandas as pd
df_ads = pd.read_csv('advertising.csv')
df_ads.head()
```

```python
# 绘图库
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False #解决中文乱码等问题
# 对所有标签和特征两两显示其相关的热力图
sns.heatmap(df_ads.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
# 显示销售额和各种广告投放金额的散点图
sns.pairplot(df_ads,
            x_vars=['微信', '微博', '其他'],
            y_vars='销售额',
            height=4, aspect=1, kind='scatter')
plt.show()
```

```python
x = np.array(df_ads.微信)  # 特征集
y = np.array(df_ads.销售额)  # 销售集
print("张量x的阶：", x.ndim)
print("张量x的形状", x.shape)
print("张量x的内容", x)
```

```python
x = x.reshape((len(x), 1))
y = y.reshape((len(y), 1))
print("张量x的阶：", x.ndim)
print("张量x的形状", x.shape)
print("张量x的内容", x)
```

```python
# 将数据集进行80％（训练集）和20％（测试集）的分割
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# 获取反向归一化数据
x_train_max = max(x_train)[0]
x_train_min = min(x_train)[0]
y_train_max = max(y_train)[0]
y_train_min = min(y_train)[0]
```

# 对数据进行归一化

```python
# 定义归一化函数
def scaler(train, test):
    minNum = train.min(axis=0)
    maxNum = train.max(axis=0)
    gap = maxNum - minNum
    train -= minNum
    train /= gap
    test -= minNum
    test /=gap
    return train, test
```

```python
x_train, x_test = scaler(x_train, x_test)
y_train, y_test = scaler(y_train, y_test)
```

```python
plt.plot(x_train, y_train, 'r.', label='训练数据')
plt.xlabel("微信")
plt.ylabel("销售额")
plt.legend()  # 显示图例
plt.show()
```

```python
# 均方误差函数
def loss_function(x, y, weight, bias):
    y_hat = weight*x + bias
    loss = y_hat - y
    cost = np.sum(loss**2)/(2*len(x))
    return cost
```

# 采用梯度下降法计算权重和偏置

```python
# 梯度下降法
def gradient_descent(x, y, w, b, lr, iter):
    l_history = np.zeros(iter)  # 初始化记录梯度下降过程中损失的数组
    w_history = np.zeros(iter)  # 初始化记录梯度下降过程中权重的数组
    b_history = np.zeros(iter)  # 初始化记录梯度下降过程中偏置的数组
    for i in range(iter):
        y_hat = w * x + b
        loss = y_hat - y
        derivative_w = x.T.dot(loss)/ len(x)  # 对权重求导
        derivative_b = sum(loss) * 1 / len(x) # 对偏导求导
        w = w - lr * derivative_w  # 结合学习速率alpha更新权重
        b = b - lr * derivative_b  # 结合学习速率alpha更新偏置
        l_history[i] = loss_function(x, y, w, b)
        w_history[i] = w
        b_history[i] = b
    return l_history, w_history, b_history
```

# 设定参数初始值

```python
iterations = 10000 # 初始迭代10000次
alpha = 0.05
weight = -5
bias = 3
print("当前损失：", loss_function(x_train, y_train, weight, bias))
```

```python
plt.plot(x_train, y_train, 'r.', label='训练数据')
line_x = np.linspace(x_train.min(), x_train.max(),500)  # x值域
line_y = [weight*xx + bias for xx in line_x]  # y_hat
plt.plot(line_x, line_y, 'b--', label="当前假设函数")  # 显示当前假设函数
plt.xlabel("微信")
plt.ylabel("销售额")
plt.legend()
plt.show()
```

```python
# 开始训练
loss_history, weight_history, bias_history = gradient_descent(x_train, y_train, weight, bias, alpha, iterations)
```

```python
# 显示损失曲线
plt.plot(loss_history, 'g--', label="损失曲线")
plt.xlabel("迭代次数")
plt.ylabel("损失")
plt.legend()
plt.show()
```

```python
# 绘制当前函数模型
plt.plot(x_train, y_train,  'k.', label="训练数据")
line_x = np.linspace(x_train.min(), x_train.max(),500)  # x值域
line_y = [weight_history[-1]*xx + bias_history[-1] for xx in line_x]  # y_hat
plt.plot(line_x, line_y, 'b--', label="当前假设函数")  # 显示当前假设函数
plt.xlabel("微信")
plt.ylabel("销售额")
plt.legend()
plt.show()
```

```python
print("当前损失",loss_function(x_train, y_train, weight_history[-1], bias_history[-1]))
print("当前权重",weight_history[-1])
print("当前偏置",bias_history[-1])
```

```python
print("测试集损失",loss_function(x_test, y_test, weight_history[-1], bias_history[-1]))
```

# 预计商品销售额

```python
x_plan = 300  # 微信投入金额为300元
x_plan = (x_plan -x_train_min)/(x_train_max - x_train_min)  # 对输入数据进行缩放
y_plan = np.dot(weight_history[-1], x_plan) + bias_history[-1]
```
